```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
data = pd.read_csv('lsd_math.csv')
```

```python
print(data)
```

```python
Time = data[['Time_Delay_in_Minutes']]
LSD = data [['LSD_ppm']]
Score = data[['Avg_Math_Test_Score']]
```

```python
type(Score)
```

```python
%matplotlib inline

plt.title('Tissue Concentration Of LSD Over Time', color='g', fontsize = 17)
plt.xlabel('Time In Minutes', color='blue', fontsize = 14)
plt.ylabel('Tissue LSD_ppm',color='blue',  fontsize= 14)
plt.text( x = -1, y = -0.5, s= 'wagner et al.(1968)', fontsize = 12)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.ylim(1,7)
plt.xlim(0, 500 )
plt.style.use('fivethirtyeight')




plt.plot(Time, LSD, color='r', linewidth = 2)
plt.show()
```

```python
regr = LinearRegression()
regr.fit(LSD, Score)
regr.coef_[0][0]
predicted_score = regr.predict(LSD)
```

```python
regr.intercept_[0]
```

```python
regr.score(LSD, Score)
```

Now create a plot for the above data 


```python
%matplotlib inline



plt.title('ARITHMETIC VS LSD-25', color='Red', fontsize = 17)
plt.xlabel('Tissue LSD ppm', fontsize = 14)
plt.ylabel('Performance Score', fontsize = 14)
plt.xlim(1, 6.5)
plt.ylim(25, 85)
plt.style.use('bmh')


plt.scatter(LSD, Score, color='blue', s=100, alpha=0.7)
plt.plot(LSD, predicted_score, color = 'cyan', linewidth=3)
plt.show()

```
